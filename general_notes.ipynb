# Recommendations 

## Memory

Currently only 8h of data can be stored. That is because we can only access the main Renesas chip, which has very little memory (32kb). This can be improved in three ways: either adding external memory to the system; add another processor that is responsible for the data storage; or access the memory of the existing Espressif chip, which is the best solution.
